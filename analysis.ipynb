```python
import pandas as pd
from datetime import datetime as dt

df = pd.read_csv(f'output.csv')
```

```python
df['is_weekend'] = df['date'].apply(lambda x: 'WEEKEND' if dt.strptime(x, '%Y-%m-%d').weekday() in [5, 6] else 'WEEKDAY')
df.insert(1, 'is_weekend', df.pop('is_weekend'))
df
```

```python
# split df into 2 data frames, one for weekdays and one for weekends
df_weekday = df[df['is_weekend'] == 'WEEKDAY']
df_weekend = df[df['is_weekend'] == 'WEEKEND']

df_weekday
```

```python
df_weekend
```

```python
df_all = df.iloc[:,2:]
df_all
```

```python
df_all_mean = df_all.mean(axis=0)
df_all_mean.head(30)
```

```python
df_weekday_mean = df_weekday.iloc[:,2:].mean(axis=0)
df_weekday_mean.head(30)
```

```python
df_weekend_mean = df_weekend.iloc[:,2:].mean(axis=0)
df_weekend_mean.head(30)
```

```python
def plot_usage(df, title):
    df.plot(kind='bar', title=title, xlabel='Hour', ylabel='Usage (kWh)')
```

```python
plot_usage(df_all_mean, title='Average Daily Usage')
```

```python
plot_usage(df_weekday_mean, title='Average Daily Usage: Weekdays')
```

```python
plot_usage(df_weekend_mean, title='Average Daily Usage: Weekends')
```

```python
TOU_PRICING = [
    (0, 8.7),
    (7, 18.2),
    (11, 12.2),
    (17, 18.2),
    (19, 8.7)
]

ULO_WEEKDAY_PRICING = [
    (0, 2.8),
    (7, 12.2),
    (16, 28.6),
    (21, 12.2),
    (23, 2.8),
]

ULO_WEEKEND_PRICING = [
    (0, 2.8),
    (7, 8.7),
    (23, 2.8),
]

TIERED_PRICING = [
    (0, 10.3),
    (750, 12.5),
]

WEEKS_IN_MONTH = 365 / 7 / 12
DAYS_IN_MONTH = 365 / 12

def get_tou_price(hour):
    for start, price in TOU_PRICING.__reversed__():
        if hour >= start:
            return price
            
def get_ulo_price(hour, is_weekend):
    if is_weekend:
        for start, price in ULO_WEEKEND_PRICING.__reversed__():
            if hour >= start:
                return price
    else:
        for start, price in ULO_WEEKDAY_PRICING.__reversed__():
            if hour >= start:
                return price
            
def get_tiered_price(usage):
    if usage < 750:
        return TIERED_PRICING[0][1]
    else:
        return TIERED_PRICING[1][1]
```

```python
day_tou_costs = map(lambda x: get_tou_price(x[0]) * x[1], enumerate(df_all_mean.values))
day_tou_cost = sum(day_tou_costs)
day_tou_cost = day_tou_cost * 0.01
day_tou_cost = day_tou_cost * DAYS_IN_MONTH

weekday_ulo_costs = map(lambda x: get_ulo_price(x[0], False) * x[1], enumerate(df_weekday_mean))
weekday_ulo_cost = sum(weekday_ulo_costs) * 0.01
weekend_ulo_costs = map(lambda x: get_ulo_price(x[0], True) * x[1], enumerate(df_weekend_mean))
weekend_ulo_cost = sum(weekend_ulo_costs) * 0.01
week_ulo_cost = (weekday_ulo_cost*5)+(weekend_ulo_cost*2)
month_ulo_cost = week_ulo_cost * WEEKS_IN_MONTH

tiered_daily_usage = sum(df_all_mean)
tiered_month_usage = tiered_daily_usage * DAYS_IN_MONTH
tiered_month_cost = tiered_month_usage * get_tiered_price(tiered_month_usage) * 0.01

print(f'TOU Cost: ${day_tou_cost:.2f}')
print(f'ULO Cost: ${month_ulo_cost:.2f}')
print(f'Tiered Cost: ${tiered_month_cost:.2f}')
```
